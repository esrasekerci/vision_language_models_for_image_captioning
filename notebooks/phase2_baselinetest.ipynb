```python
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt

# Load predictions (which already has reference captions)
test_preds_df = pd.read_csv("/kaggle/input/di725-project-phase2/test_preds.csv")

# Pick a test example (e.g., the first row)
idx = 15
row = test_preds_df.iloc[idx]
image_name = row['image']
generated_caption = row['generated_caption']
reference_caption = row['reference_caption']  # already in your CSV

# Image path
img_path = f"/kaggle/input/di725-dataset/resized/{image_name}"
img = Image.open(img_path)

# Save figure
plt.figure(figsize=(4, 4))
plt.imshow(img)
plt.axis('off')
plt.tight_layout()
plt.savefig("example_image_caption.png", bbox_inches='tight', dpi=300)
plt.close()

# Now write the captions to a text file
with open("example_captions.txt", "w") as f:
    f.write("Generated Caption: " + generated_caption + "\n")
    f.write("Reference Caption: " + reference_caption + "\n")
```
